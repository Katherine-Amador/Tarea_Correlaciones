```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
wh = pd.read_csv('2018.csv')
```

```python
wh
```

```python
wh.describe()
```

```python
wh.columns
```

```python
columns_to_use=['Score', 'GDP per capita',
       'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
```

```python
corr_use=wh.loc[: , columns_to_use].corr()
```

```python
corr_use
```

```python
plt.matshow(corr_use)
```

```python
f, ax = plt.subplots(figsize=(10,8))

sns.heatmap(corr_use, mask=np.zeros_like(corr_use, dtype=np.bool), cmap=sns.diverging_palette(280, 50,  as_cmap=True),
           square=True, ax=ax)
```
